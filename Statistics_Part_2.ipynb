#Ques1. What is hypothesis testing in statistics?
**Ans:**A method used to make inferences or decisions about a population based on sample data.



#Ques2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
**Ans:** **Null Hypothesis (H₀):** Assumes no effect or no difference.

**Alternative Hypothesis (H₁):** Assumes there is an effect or difference.
They are mutually exclusive; rejecting one supports the other.

#Ques3. What is the significance level in hypothesis testing, and why is it important?
**Ans:**The significance level (α) is the threshold for rejecting H₀, typically set at 0.05.
It represents the probability of a Type I error (false positive).



#Ques4. What does a P-value represent in hypothesis testing?
**Ans:**The P-value is the probability of observing the sample data (or more extreme) assuming H₀ is true.

#Ques5. How do you interpret the P-value in hypothesis testing?
**Ans:**
If P ≤ α → Reject H₀ (significant result)

If P > α → Fail to reject H₀ (not significant)

#Ques6. What are Type 1 and Type 2 errors in hypothesis testing?
**Ans:**
Type I Error: Rejecting a true H₀ (false positive)

Type II Error: Failing to reject a false H₀ (false negative)

#Ques7. What is the difference between a one-tailed and a two-tailed test?
**Ans:**One-tailed: Tests effect in one direction (e.g., > or <)

Two-tailed: Tests for any difference (e.g., ≠)



#Ques8. What is the Z-test, and when is it used?
**Ans:**A Z-test is used when the population standard deviation is known and the sample size is large (n ≥ 30).

#Ques9. How do you calculate the Z-score, and what does it represent?
**Ans:**
Z=(x(Bar)−μ)/(σ/ Root (n))
It shows how many standard errors the sample mean is from the population mean.



#Ques10. What is the T-distribution, and when should it be used?
**Ans:**The T-distribution is used when the sample size is small (n < 30) and the population standard deviation is unknown.

#Ques11. What is the difference between a Z-test and a T-test?
**Ans:**Z-test: Known population standard deviation

T-test: Unknown population standard deviation, uses sample std. dev and T-distribution

#Ques 12. What is the T-test, and how is it used in hypothesis testing?
**Ans:**The T-test compares the means of one or two groups when the population standard deviation is unknown. It helps determine if the mean difference is statistically significant.


#Ques 13. What is the relationship between Z-test and T-test in hypothesis testing?
**Ans:** Both test mean differences.

Use Z-test when population std dev is known and sample size is large.

Use T-test when std dev is unknown or sample size is small.

#Ques 14. What is a confidence interval, and how is it used to interpret statistical results?
**Ans:**A confidence interval (CI) provides a range in which the population parameter is likely to fall, with a certain level of confidence (e.g., 95%).



#Ques 15. What is the margin of error, and how does it affect the confidence interval?
**Ans:**The margin of error is the range above and below the sample estimate. A larger margin makes the CI wider, indicating more uncertainty.



#Ques 16. How is Bayes' Theorem used in statistics, and what is its significance?
**Ans:**Bayes' Theorem updates the probability of a hypothesis based on new evidence.
It is important in probabilistic reasoning and machine learning.



#Ques 17. What is the Chi-square distribution, and when is it used?
**Ans:**A Chi-square distribution is used for tests involving categorical data, especially to test independence or goodness of fit.



#Ques 18. What is the Chi-square goodness of fit test, and how is it applied?
**Ans:**It tests whether observed frequencies match expected frequencies for a categorical variable.
If the difference is large, the null hypothesis (good fit) is rejected.



#Ques 19. What is the F-distribution, and when is it used in hypothesis testing?
**Ans:**The F-distribution is used to compare two variances. It's also the basis of the ANOVA test.

#Ques 20. What is an ANOVA test, and what are its assumptions?
**Ans:**
ANOVA (Analysis of Variance) compares the means of three or more groups.
Assumptions:

- Independence

- Normality

- Equal variances (homogeneity of variance)

#Ques 21. What are the different types of ANOVA tests?
**Ans:**
- One-Way ANOVA: One independent variable (factor)

- Two-Way ANOVA: Two factors

- Repeated Measures ANOVA: Same subjects tested multiple times



#Ques 22. What is the F-test, and how does it relate to hypothesis testing?
**Ans:**
 The F-test compares variances to determine if they are significantly different. It’s used in ANOVA and regression analysis.

#Ques2.  Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define range for the discrete uniform distribution (e.g., 1 to 6 for a die)
low = 1   # inclusive
high = 7  # exclusive in scipy.stats.randint

# Create distribution object
dist = randint(low, high)

# Generate x-values (possible outcomes)
x = np.arange(low, high)

# Calculate PMF
pmf = dist.pmf(x)

# Plot PMF
plt.figure(figsize=(8, 5))
plt.stem(x, pmf, basefmt=" ")
plt.title("Discrete Uniform Distribution PMF (1 to 6)")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks(x)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

#Ques3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

```python
from scipy.stats import bernoulli

def bernoulli_pdf(p):
    """
    Calculate and return the PMF of a Bernoulli distribution with success probability p.

    Parameters:
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    dict: A dictionary with keys 0 and 1 and their corresponding probabilities
    """
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1")

    dist = bernoulli(p)
    return {0: dist.pmf(0), 1: dist.pmf(1)}

# Example usage:
probability = 0.7
pmf = bernoulli_pdf(probability)
print("Bernoulli PMF with p =", probability)
print("P(X = 0):", pmf[0])
print("P(X = 1):", pmf[1])
```

#Ques 9. Write a Python function to calculate Z-scores from a dataset and plot them?

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculate Z-scores for a dataset and plot a histogram of the Z-scores.

    Parameters:
    data (array-like): Numerical dataset
    """
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot histogram of Z-scores
    plt.figure(figsize=(8, 5))
    plt.hist(z_scores, bins=30, color='skyblue', edgecolor='black', density=True)
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z = 0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=50, scale=10, size=1000)
z_scores = calculate_and_plot_z_scores(sample_data)
```

#Ques14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_score_plot(data, x_value):
    """
    Calculates the Z-score for a given x_value from data,
    and compares it to the standard normal distribution.
    """
    # Step 1: Compute mean and standard deviation
    mean = np.mean(data)
    std = np.std(data, ddof=1)

    # Step 2: Compute Z-score
    z = (x_value - mean) / std
    print(f"Data Mean = {mean:.2f}, Standard Deviation = {std:.2f}")
    print(f"Z-score for X = {x_value} is {z:.2f}")

    # Step 3: Plot Standard Normal Distribution
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, color='blue', label='Standard Normal PDF')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')
    plt.title("Z-score Comparison to Standard Normal Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Example usage:
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)  # Simulated dataset
x_value = 65  # Example data point
z_score_plot(data, x_value)
```

#Qeus15. Implement hypothesis testing using Z-statistics for a sample dataset?

```python
import numpy as np
from scipy.stats import norm

# Known parameters
mu = 170         # Population mean
sigma = 10       # Population std dev
alpha = 0.05     # Significance level
n = 50           # Sample size

# Step 1: Generate a sample (simulate data)
np.random.seed(42)
sample = np.random.normal(loc=172, scale=sigma, size=n)
sample_mean = np.mean(sample)

# Step 2: Calculate Z-score
z = (sample_mean - mu) / (sigma / np.sqrt(n))

# Step 3: Determine critical Z-value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Step 4: Hypothesis Testing Result
print(f"Sample Mean = {sample_mean:.2f}")
print(f"Z-score = {z:.4f}")
print(f"Z-critical (±{z_critical:.4f})")

if abs(z) > z_critical:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

#Ques16. Create a confidence interval for a dataset using Python and interpret the result?

```python
import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate confidence interval for the mean of a dataset.
    """
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)

    z_critical = norm.ppf(1 - (1 - confidence) / 2)  # two-tailed z value
    margin_of_error = z_critical * std_err

    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
    return (lower_bound, upper_bound)

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=50)
calculate_confidence_interval(sample_data, confidence=0.95)
```

#Ques 17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?

```python
import numpy as np
from scipy.stats import norm

# Step 1: Generate data from a normal distribution
np.random.seed(1)
data = np.random.normal(loc=100, scale=15, size=50)  # mean=100, std=15, n=50

# Step 2: Calculate 95% confidence interval
def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    std_err = std_dev / np.sqrt(n)

    z_score = norm.ppf(1 - (1 - confidence) / 2)
    margin = z_score * std_err

    lower = mean - margin
    upper = mean + margin

    print(f"Sample Mean = {mean:.2f}")
    print(f"{int(confidence*100)}% Confidence Interval = ({lower:.2f}, {upper:.2f})")
    return lower, upper

# Run the function
confidence_interval(data)
```

#Ques1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform one-sample Z-test.
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    standard_error = pop_std / np.sqrt(n)

    # Calculate Z-score
    z = (sample_mean - pop_mean) / standard_error

    # Two-tailed critical Z value
    z_critical = norm.ppf(1 - alpha / 2)

    # P-value from Z-score
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Display results
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-score = {z:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Z-critical value for α = {alpha} is ±{z_critical:.4f}")

    if abs(z) > z_critical:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

# Example usage
np.random.seed(0)
sample = np.random.normal(loc=103, scale=15, size=50)  # Sample data
population_mean = 100
population_std = 15

z_test(sample, population_mean, population_std)
```

#Ques3 .Implement a one-sample Z-test using Python to compare the sample mean with the population mean?

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, pop_mean, pop_std, alpha=0.05):
    """
    Perform a one-sample Z-test.

    Parameters:
    sample_data : array-like
        Sample observations
    pop_mean : float
        Population mean to compare with
    pop_std : float
        Known population standard deviation
    alpha : float
        Significance level (default 0.05 for 95% confidence)

    Returns:
    z_score, p_value
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    se = pop_std / np.sqrt(n)

    # Calculate Z-score
    z = (sample_mean - pop_mean) / se

    # Two-tailed p-value
    p = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-score = {z:.4f}")
    print(f"P-value = {p:.4f}")

    if p < alpha:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

    return z, p

# Example usage
np.random.seed(1)
sample = np.random.normal(loc=102, scale=10, size=50)
population_mean = 100
population_std = 10

one_sample_z_test(sample, population_mean, population_std)
```

#Ques4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_and_plot(sample_data, pop_mean, pop_std, alpha=0.05):
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    standard_error = pop_std / np.sqrt(n)

    # Calculate Z-score
    z_score = (sample_mean - pop_mean) / standard_error
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Z critical values for two-tailed test
    z_crit = norm.ppf(1 - alpha / 2)

    # Print results
    print(f"Sample Mean = {sample_mean:.2f}")
    print(f"Z-score = {z_score:.4f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Z-critical (±{z_crit:.4f})")

    if abs(z_score) > z_crit:
        print("Result: Reject the null hypothesis (significant difference).")
    else:
        print("Result: Fail to reject the null hypothesis (no significant difference).")

    # Plot standard normal curve
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color="blue")

    # Shade critical regions
    plt.fill_between(x, y, where=(x <= -z_crit), color='red', alpha=0.3, label="Rejection Region (left)")
    plt.fill_between(x, y, where=(x >= z_crit), color='red', alpha=0.3, label="Rejection Region (right)")

    # Plot Z-score
    plt.axvline(z_score, color='black', linestyle='--', label=f'Z-score = {z_score:.2f}')

    # Labels and legend
    plt.title("Two-Tailed Z-Test and Decision Regions")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Example usage
np.random.seed(0)
sample = np.random.normal(loc=104, scale=10, size=40)
population_mean = 100
population_std = 10

two_tailed_z_test_and_plot(sample, population_mean, population_std)
```

#Ques 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0=100, mu_1=105, sigma=10, n=30, alpha=0.05, tail='right'):
    """
    Visualizes Type I and Type II errors for a one-tailed Z-test.
    Parameters:
        mu_0 : float - Population mean under H0
        mu_1 : float - Population mean under H1
        sigma : float - Population standard deviation
        n : int - Sample size
        alpha : float - Significance level
        tail : str - 'right' or 'left' for one-tailed test
    """
    se = sigma / np.sqrt(n)
    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)

    # H0 and H1 distributions
    h0_pdf = norm.pdf(x, mu_0, se)
    h1_pdf = norm.pdf(x, mu_1, se)

    # Z critical for one-tailed test
    if tail == 'right':
        z_crit = norm.ppf(1 - alpha)
        crit_val = mu_0 + z_crit * se
        type2_area = norm.cdf(crit_val, loc=mu_1, scale=se)
    elif tail == 'left':
        z_crit = norm.ppf(alpha)
        crit_val = mu_0 + z_crit * se
        type2_area = 1 - norm.cdf(crit_val, loc=mu_1, scale=se)
    else:
        raise ValueError("tail must be 'right' or 'left'")

    beta = type2_area
    power = 1 - beta

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(x, h0_pdf, label='H₀ Distribution (μ₀)', color='blue')
    plt.plot(x, h1_pdf, label='H₁ Distribution (μ₁)', color='green')

    # Shade Type I Error region
    if tail == 'right':
        plt.fill_between(x, 0, h0_pdf, where=(x >= crit_val), color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, 0, h1_pdf, where=(x <= crit_val), color='orange', alpha=0.3, label='Type II Error (β)')
    else:
        plt.fill_between(x, 0, h0_pdf, where=(x <= crit_val), color='red', alpha=0.3, label='Type I Error (α)')
        plt.fill_between(x, 0, h1_pdf, where=(x >= crit_val), color='orange', alpha=0.3, label='Type II Error (β)')

    plt.axvline(crit_val, color='black', linestyle='--', label=f'Critical Value = {crit_val:.2f}')
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

    print(f"Critical Value = {crit_val:.2f}")
    print(f"Type I Error (α) = {alpha}")
    print(f"Type II Error (β) = {beta:.4f}")
    print(f"Power of the Test = {power:.4f}")
```

#Ques 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance

```python
import numpy as np
from scipy.stats import t

def confidence_interval_mean(sample_data, confidence=0.95):
    """
    Calculate the confidence interval for the mean of a sample using T-distribution.

    Parameters:
        sample_data : list or array-like
            The sample data (numeric)
        confidence : float
            The confidence level (default = 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound)
    """
    n = len(sample_data)
    mean = np.mean(sample_data)
    std_dev = np.std(sample_data, ddof=1)
    se = std_dev / np.sqrt(n)

    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * se

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"Sample Mean = {mean:.2f}")
    print(f"Confidence Level = {int(confidence*100)}%")
    print(f"Confidence Interval = ({lower:.2f}, {upper:.2f})")

    return lower, upper

# Example usage:
np.random.seed(0)
sample = np.random.normal(loc=100, scale=15, size=50)
confidence_interval_mean(sample)
```

#Ques 10. Write a Python program to calculate the margin of error for a given confidence level using sample data?

```python
import numpy as np
from scipy.stats import t

def calculate_margin_of_error(sample_data, confidence=0.95):
    """
    Calculate the margin of error for a sample mean at the given confidence level.

    Parameters:
        sample_data (array-like): Sample observations
        confidence (float): Confidence level (default 0.95 for 95%)

    Returns:
        float: Margin of Error
    """
    n = len(sample_data)
    sample_std = np.std(sample_data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    # t-critical value
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)

    # Margin of Error
    margin_error = t_crit * standard_error

    print(f"Sample Size = {n}")
    print(f"Sample Std Dev = {sample_std:.4f}")
    print(f"T-critical = {t_crit:.4f}")
    print(f"Margin of Error (at {int(confidence*100)}% confidence) = ±{margin_error:.2f}")

    return margin_error

# Example usage:
np.random.seed(0)
sample = np.random.normal(loc=100, scale=15, size=50)
calculate_margin_of_error(sample, confidence=0.95)
```

#Ques 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?

```python
def bayes_theorem(prior_H, likelihood_E_given_H, likelihood_E_given_not_H):
    """
    Applies Bayes' Theorem to calculate P(H|E)
    """
    # Prior of not having the disease
    prior_not_H = 1 - prior_H

    # Total probability of evidence (positive test)
    P_E = likelihood_E_given_H * prior_H + likelihood_E_given_not_H * prior_not_H

    # Posterior: Probability of H given E
    posterior = (likelihood_E_given_H * prior_H) / P_E

    return posterior

# Define values
prior_disease = 0.01                # P(H)
true_positive = 0.99                # P(E|H)
false_positive = 0.05               # P(E|¬H)

# Compute P(H|E)
posterior_prob = bayes_theorem(prior_disease, true_positive, false_positive)

print(f"Probability of having disease given a positive test: {posterior_prob:.4f} or {posterior_prob*100:.2f}%")
```

#Ques 12. Perform a Chi-square test for independence between two categorical variables in Python?

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Step 1: Create a contingency table (2D frequency table)
# Example: Gender vs Product Preference
data = {
    'Electronics': [30, 10],  # [Male, Female]
    'Clothing': [20, 25],
    'Groceries': [10, 30]
}
table = pd.DataFrame(data, index=['Male', 'Female'])

# Step 2: Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(table)

# Step 3: Output results
print("Contingency Table:")
print(table)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=table.index, columns=table.columns))

print(f"\nChi-square Statistic = {chi2:.4f}")
print(f"Degrees of Freedom = {dof}")
print(f"P-value = {p:.4f}")

# Step 4: Interpretation
alpha = 0.05
if p < alpha:
    print("Result: Reject the null hypothesis. Variables are dependent.")
else:
    print("Result: Fail to reject the null hypothesis. Variables are independent.")
```

#Ques14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.?

```python
import numpy as np
from scipy.stats import chisquare

# Step 1: Observed frequencies from experiment (e.g., 6-sided die rolled 120 times)
observed = np.array([15, 18, 20, 17, 25, 25])  # Face 1 to 6

# Step 2: Expected frequencies (assuming fair die)
total_rolls = np.sum(observed)
expected = np.full_like(observed, total_rolls / len(observed))

# Step 3: Perform Chi-square goodness-of-fit test
chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

# Step 4: Display results
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected.astype(int))
print(f"\nChi-square Statistic = {chi2_stat:.4f}")
print(f"P-value = {p_value:.4f}")

# Step 5: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The data does NOT follow the expected distribution.")
else:
    print("Result: Fail to reject the null hypothesis. The data follows the expected distribution.")
```
