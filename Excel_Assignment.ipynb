Question 1: What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly explain with examples.

Answer:
'Paste' copies everything from the source cell (value, formula, formatting, etc.) and pastes it into the destination cell.
'Paste Special' allows you to choose what to paste — only values, only formatting, only formulas, etc.

Example:

If you copy a cell with a formula and use Paste, the formula will be pasted.

If you use Paste Special → Values, only the result of the formula will be pasted.

Question 2: Describe the functions and usefulness of 'Freeze Panes' and 'Split Panes' in Excel.

Answer:

Freeze Panes: Locks specific rows or columns so they remain visible while scrolling through the rest of the worksheet. Useful for keeping headers visible.

Split Panes: Divides the window into separate scrollable areas, allowing you to view different parts of the same sheet simultaneously.

Question 3: Explain the difference between inserting a new row and inserting a new column in Excel. Can you insert multiple rows or columns at once?

Answer:

Inserting a row adds a new horizontal set of cells above the selected row.

Inserting a column adds a new vertical set of cells to the left of the selected column.
Yes, you can insert multiple rows or columns at once by selecting the same number of existing rows or columns before inserting.

Question 4: What are logical functions in Excel? Provide examples of at least two logical functions and their applications.

Answer:
Logical functions are used to test conditions and return TRUE or FALSE.
Examples:

IF: Checks a condition and returns one value if true, another if false.
Example: =IF(A1>50, "Pass", "Fail")

AND: Returns TRUE only if all conditions are true.
Example: =AND(A1>50, B1<100)

Question 5: Discuss the purpose of 'XLOOKUP' and how it differs from the traditional 'VLOOKUP' function.

Answer:
XLOOKUP is used to search for a value in a range and return a corresponding value from another range.
Differences from VLOOKUP:

XLOOKUP can search in any direction (left, right, up, down).

It doesn’t require a column index number.

It allows a default value if the lookup value is not found.

It works with vertical and horizontal data.
